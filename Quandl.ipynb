# Build a simple trading strategy

```python
# prompt: using quandl get the latest data , add columns moving avg column 10, and 50 ,  ma10 and ma 50

# Get the latest data (replace 'YOUR_DATASET_CODE' with the actual Quandl code)
data = quandl.get("WIKI/AAPL", start_date='2013-12-31', end_date='2024-08-15')

# Calculate moving averages
data['ma10'] = data['Close'].rolling(10).mean()
data['ma50'] = data['Close'].rolling(50).mean()

print(data.tail())
```

```python
data['Share'] = [1 if data.loc[ei, 'ma10'] > data.loc[ei, 'ma50'] else -1 for ei in data.index]

data.head()
```

```python
#Close previous data
filtered_data = data.loc['2015-01-01':'2018-12-31']
filtered_data['Close1'] = filtered_data['Close'].shift(-1)

filtered_data['Profit'] = [filtered_data.loc[ei, 'Close1'] - filtered_data.loc[ei, 'Close'] if filtered_data.loc[ei, 'Share'] == 1 else 0 for ei in filtered_data.index]

filtered_data['Profit'].plot()
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```

```python
filtered_data['Wealth'] = filtered_data['Profit'].cumsum()

filtered_data['Wealth'].plot()

plt.title('Total money you win is {:.2f}'.format(filtered_data.loc[filtered_data.index[-2], 'Wealth']))
plt.show()
```
