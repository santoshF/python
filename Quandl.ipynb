````python
import numpy as np

my_data = np.array([22, 15, 53, 6, 36])
x_axis = np.arange(len(my_data))
print(x_axis)
````

````python
my_mean = my_data.mean()
my_stdev = my_data.std()
````

````python
my_data_norm = (my_data - my_mean)/my_stdev
print(my_data_norm)
my_data_domain_standardized = (my_data - my_data.min())/(my_data.max() - my_data.min())
print(my_data_domain_standardized)

fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(10,5))

ax1.plot(x_axis,my_data)
ax2.plot(x_axis,my_data_norm)
ax3.plot(x_axis,my_data_domain_standardized)

plt.show()
````

````python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/santoshF/python/master/happyscore_and_income.csv')
data.head()

````

````python
# prompt: Using dataframe data: scatter plot with avg_income above 15000 and label the largest and smallest using matplotlib.pyplot library

import matplotlib.pyplot as plt

# Filter data for avg_income above 15000
filtered_data = data[data['avg_income'] > 15000]

# Create scatter plot
plt.scatter(filtered_data['avg_income'], filtered_data['happyScore'])
plt.xlabel('Average Income')
plt.ylabel('Happy Score')

# Find and label largest and smallest points
largest = filtered_data.loc[filtered_data['avg_income'].idxmax()]
smallest = filtered_data.loc[filtered_data['avg_income'].idxmin()]
plt.text(largest['avg_income'], largest['happyScore'], largest['country'])
plt.text(smallest['avg_income'], smallest['happyScore'], smallest['country'])

plt.show()
````

````python
dataframe = pd.read_csv('https://raw.githubusercontent.com/santoshF/python/master/Banknote_authentication_dataset.csv')

dataframe.head()
````

````python
# prompt: Using dataframe dataframe:  run K-means on the given dataset, visualise the results,  re-run K-means several times and visualise the results. could we add all the rerun in a chart use matplotlib.pyplot , limit rerun charts to 3

import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Run K-means and visualize results
kmeans = KMeans(n_clusters=3)
kmeans.fit(dataframe)
labels = kmeans.labels_
plt.scatter(dataframe['V1'], dataframe['V2'], c=labels)
plt.title('K-means Clustering')
plt.show()

# Re-run K-means and visualize results (limited to 3 reruns)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i in range(3):
    kmeans = KMeans(n_clusters=3, random_state=i)  # Different random state for each rerun
    kmeans.fit(dataframe)
    labels = kmeans.labels_
    axes[i].scatter(dataframe['V1'], dataframe['V2'], c=labels)
    axes[i].set_title(f'K-means Clustering (Rerun {i+1})')
plt.tight_layout()
plt.show()
````
