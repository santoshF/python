```python
import numpy as np

x = np.array([2,3,5,6])
y = np.array([-8,-7,15,16])

my_range = np.arange(4)
print(my_range)

mean_x = np.mean(x)
mean_y = np.mean(y)

print(mean_x)
print(mean_y)

fig , axs = plt.subplots(1,2, sharey=True)

axs[0].stem(my_range,x, label='x values', bottom=mean_x)
axs[1].stem(my_range,y, label='y values', bottom=mean_y)

lengend_x = axs[0].legend(['x','y'])
lengend_y = axs[1].legend(['x','y'])

plt.show()
```

```python
import numpy as np

my_data = np.array([22, 15, 53, 6, 36])
x_axis = np.arange(len(my_data))
print(x_axis)
```

```python
my_mean = my_data.mean()
my_stdev = my_data.std()
```

```python
my_data_norm = (my_data - my_mean)/my_stdev
print(my_data_norm)
my_data_domain_standardized = (my_data - my_data.min())/(my_data.max() - my_data.min())
print(my_data_domain_standardized)

fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(10,5))

ax1.plot(x_axis,my_data)
ax2.plot(x_axis,my_data_norm)
ax3.plot(x_axis,my_data_domain_standardized)

plt.show()
```
