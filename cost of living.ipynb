```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock symbol
symbol = "ETN"

# Download historical data
etn_data = yf.download(symbol, period="5y")  # Get data for the last 5 years

# Basic analysis

# Calculate daily returns
etn_data['Daily Return'] = etn_data['Adj Close'].pct_change()

# Visualize closing price
plt.figure(figsize=(12, 6))
plt.plot(etn_data['Adj Close'], label='Adjusted Close Price')
plt.title(f'{symbol} Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Visualize daily returns
plt.figure(figsize=(12, 6))
plt.plot(etn_data['Daily Return'], label='Daily Return')
plt.title(f'{symbol} Daily Returns')
plt.xlabel('Date')
plt.ylabel('Return')
plt.legend()
plt.show()
plt.savefig('ETN_closing_prices.png')
# You can further analyze the data by:
# - Calculating moving averages
# - Identifying trends and patterns
# - Analyzing volatility
# - Comparing with other stocks
# - Using technical indicators
```

```python
# Plot trading volume
plt.figure(figsize=(10, 6))
plt.bar(etn_data.index, etn_data['Volume'], label='Volume')
plt.title('ETN Trading Volume')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.show()
plt.savefig('ETN_trading_volume.png')
```

```python
# Calculate moving averages
etn_data['MA20'] = etn_data['Close'].rolling(window=20).mean()
etn_data['MA50'] = etn_data['Close'].rolling(window=50).mean()

# Plot moving averages
plt.figure(figsize=(10, 6))
plt.plot(etn_data['Close'], label='ETN Close Price')
plt.plot(etn_data['MA20'], label='20-Day MA')
plt.plot(etn_data['MA50'], label='50-Day MA')
plt.title('ETN Closing Prices and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
plt.savefig('ETN_moving_averages.png')
```

```python
# prompt: import a in data frame data from /content/cost-of-living_v2.csv and show visuals

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV file
df = pd.read_csv('/content/cost-of-living_v2.csv')


# Visualizations
df['Rent per Month (1 Bedroom)'] = df['x48']

# 1. Histogram of Rent Prices
plt.figure(figsize=(10, 6))
plt.hist(df['Rent per Month (1 Bedroom)'], bins=20)
plt.title('Distribution of Rent Prices (1 Bedroom)')
plt.xlabel('Rent Price')
plt.ylabel('Frequency')
plt.show()

# 2. Box plot of Meal Costs
df['Meal, Inexpensive Restaurant'] = df['x3']
plt.figure(figsize=(10, 6))
sns.boxplot(x='Meal, Inexpensive Restaurant', y='country', data=df)
plt.title('Meal Costs at Inexpensive Restaurants by Country')
plt.xlabel('Meal Cost')
plt.ylabel('Country')
plt.show()



# 3. Scatter plot of Rent vs. Meal Costs
plt.figure(figsize=(10, 6))
plt.scatter(df['Rent per Month (1 Bedroom)'], df['Meal, Inexpensive Restaurant'])
plt.title('Rent vs. Meal Costs')
plt.xlabel('Rent per Month (1 Bedroom)')
plt.ylabel('Meal, Inexpensive Restaurant')
plt.show()

# 4. Bar chart of Average Rent by Country
average_rent_by_country = df.groupby('country')['Rent per Month (1 Bedroom)'].mean()
plt.figure(figsize=(10, 6))
average_rent_by_country.plot(kind='bar')
plt.title('Average Rent by Country')
plt.xlabel('Country')
plt.ylabel('Average Rent')
plt.show()

# 5. Heatmap of Correlation between numerical columns
# Calculate correlation matrix for numerical columns only
correlation_matrix = df.select_dtypes(include=['number']).corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# You can customize these visualizations further to explore the data more deeply.
# For instance, you can filter the data based on specific countries or cities,
# and create different types of charts to answer specific questions.
```
